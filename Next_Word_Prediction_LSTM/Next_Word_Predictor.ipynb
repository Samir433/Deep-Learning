```python
# import required libraries
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([docs])
```

```python
tokenizer.word_index
```

```python
for sentence in docs.split('\n'):
  print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
# create dataset such that the previous words are the input for the next word
input_list = []
for sentence in docs.split('\n'):
  tokenized_word = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_word)):
    input_list.append(tokenized_word[:i+1])
```

```python
input_list
```

```python
size = [len(i) for i in input_list]
```

```python
max(size)
```

```python
# create datset apply prefix padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
pad = pad_sequences(input_list, 40, padding='pre')
```

```python
pad
```

```python
X = pad[:,:-1]
# last column of each row is our next word
y = pad[:,-1]
```

```python
X
```

```python
y
```

```python
word = tokenizer.word_index
```

```python
# total number of the word in the docs
len(word)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y, num_classes=396)
```

```python
y.shape
```

```python
X.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM
```

```python
model = Sequential()
model.add(Embedding(396, 100, input_length=39))
model.add(LSTM(150))
model.add(Dense(396,activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
summary = model.fit(X, y, epochs=100)
```

```python
import matplotlib.pyplot as plt
plt.plot(summary.history['accuracy'])
```

```python
word = "journey of a thousand"
# first step -->> find the token
token_word = tokenizer.texts_to_sequences([word])[0]
# second step = padding
word_after_padding = pad_sequences([token_word], 39, padding='pre')

predicted_word_token = np.argmax(model.predict(word_after_padding))

for word, index in tokenizer.word_index.items():
  if index == predicted_word_token:
    print(word)
```

```python
def predict_next_word(word):
  # first step -->> find the token
  # predict next 6 word after that
  for i in range(6):
    token_word = tokenizer.texts_to_sequences([word])[0]
    word_after_padding = pad_sequences([token_word], 10, padding='pre')
    predicted_word_token = np.argmax(model.predict(word_after_padding))

    for curr_word, index in tokenizer.word_index.items():
      if index==predicted_word_token:
        print(word +" " + curr_word)
        word = word +" " + curr_word
        time.sleep(2)
```
